### Standarize data Format

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Order Date']
```

```python
df['Month'] = df['Order Date'].dt.month
df['quarter'] = df['Order Date'].dt.to_period("Q")
```

```python
df['quarter']
```

### Exploratory Data Analysis

```python
monthly_sales = df.groupby('Month')['Sales'].sum()
monthly_sales
```

```python
plt.figure(figsize = (10,5))
monthly_sales.plot(kind="bar")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel('Sales Revenue')
plt.show()
```

```python
region_sales = df.groupby('City')['Sales'].sum()
region_sales.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Revenue')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Example: Forecast next month's sales
df_monthly = df.groupby('Month').agg({'Sales':'sum'})  # Assuming 1 year data
X = np.array(df_monthly.index).reshape(-1, 1)
y = df_monthly['Sales'].values
model = LinearRegression().fit(X, y)
next_month = np.array([[13]])
predicted_sales = model.predict(next_month)
print(f"Predicted sales for next month: {predicted_sales[0]}")
```
